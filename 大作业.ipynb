```python
from tkinter import*
import tkinter.messagebox as messagebox
from tkintertable import TableCanvas, TableModel
import pandas as pd 
import numpy as np 
from sklearn import linear_model
import matplotlib.pyplot as plt
from matplotlib.figure import Figure 
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
from PIL import ImageGrab
import pmdarima as pmd
%matplotlib inline 
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
def auto_input_file():
    global SH
    SH = pd.read_csv("F:/大二.5下/课程/大计基/2020大计基/大作业/shanghai_index.csv")

def manu_input_file():
    file_name = text1.get(1.0,END)
    global SH
    SH = pd.read_csv("F:/大二.5下/课程/大计基/2020大计基/大作业/"+ str(file_name[0:-1]))

def vis_table():
    frm = Frame()
    table = TableCanvas(frm)
    table.importCSV("F:/大二.5下/课程/大计基/2020大计基/大作业/shanghai_index.csv")
    table.show()
    frm.pack(fill=BOTH, expand=1)

def draw_total():
    f_plot.clear()
    f_plot.plot(pd.to_datetime(SH.iloc[:,0]), SH.iloc[:,1])
    canvas.draw()

def draw_LR():
    f_plot.clear()
    regr = linear_model.LinearRegression()
    x = np.array(SH.iloc[:,1]).reshape(-1,1)
    y = np.array(SH.iloc[:,2]).reshape(-1,1)
    regr.fit(x, y)
    f_plot.scatter(x, y, color = 'y')
    xreg = np.linspace(2600, 3600, 10).reshape(-1,1)
    f_plot.plot(xreg, regr.predict(xreg), color = 'red', lw = 3)
    f_plot.set_xlabel("成交量")
    f_plot.set_ylabel("股价")
    canvas.draw()

def draw_revenue():
    f_plot.clear()
    SH["Revenue"] = (SH["Price"]-SH["Price"].shift(1))/SH["Price"].shift(1)
    f_plot.hist(SH["Revenue"])
    canvas.draw()

# 保存当前图片
def save_fig():
    f.savefig("F:/大二.5下/课程/大计基/2020大计基/大作业/pic.png")

def arima_7():
    f_plot.clear()
    model = pmd.auto_arima(SH["Price"][-120:], seasonal = True, m = 12, n_jobs = 2, suppress_warnings = True)
    forecasts = model.predict(7)
    f_plot.plot(pd.to_datetime(SH.iloc[:,0][-120:]), SH.iloc[:,1][-120:])
    f_plot.plot(pd.date_range(start='2020-03-12', periods = 7), forecasts, lw = 3)
    canvas.draw()

def arima_14():
    f_plot.clear()
    model = pmd.auto_arima(SH["Price"][-120:], seasonal = True, m = 12, n_jobs = 2, suppress_warnings = True)
    forecasts = model.predict(14)
    f_plot.plot(pd.to_datetime(SH.iloc[:,0][-120:]), SH.iloc[:,1][-120:])
    f_plot.plot(pd.date_range(start='2020-03-12', periods = 14), forecasts, lw = 3)
    canvas.draw()

def arima_28():
    f_plot.clear()
    model = pmd.auto_arima(SH["Price"][-120:], seasonal = True, m = 12, n_jobs = 2, suppress_warnings = True)
    forecasts = model.predict(28)
    f_plot.plot(pd.to_datetime(SH.iloc[:,0][-120:]), SH.iloc[:,1][-120:])
    f_plot.plot(pd.date_range(start='2020-03-12', periods = 28), forecasts, lw = 3)
    canvas.draw()
```

```python
def describe():
    top_des = Tk()
    top_des.geometry('800x800')
    top_des.title("描述性统计")

    global f
    f = Figure(figsize = (8,6), dpi = 100)
    global f_plot
    f_plot = f.add_subplot(111)
    global canvas
    canvas = FigureCanvasTkAgg(f, top_des)

    label_1 = Label(top_des, text="平均数: " + str(SH.iloc[:,1].mean()))
    label_2 = Label(top_des, text="众数: " + str(np.argmax(np.bincount(SH.iloc[:,1]))))
    label_3 = Label(top_des, text="中位数: " + str(SH.iloc[:,1].median()))
    label_4 = Label(top_des, text="下四分位数: " + str(np.percentile(SH.iloc[:,1], 25)))
    label_5 = Label(top_des, text="上四分位数: " + str(np.percentile(SH.iloc[:,1], 75)))
    button_4 = Button(top_des, text = "A股总体趋势", width = 15, height = 2, command = draw_total)
    button_5 = Button(top_des, text = "股价与成交量回归分析", width = 15, height = 2, command = draw_LR)
    button_6 = Button(top_des, text = "A股收益率统计", width = 15, height = 2, command = draw_revenue)
    button_7 = Button(top_des, text = "保存图片", width = 15, height = 2, command = save_fig)
    button_8 = Button(top_des, text = "7天预测", width = 15, height = 2, command = arima_7)
    button_9 = Button(top_des, text = "14天预测", width = 15, height = 2, command = arima_14)
    button_10 = Button(top_des, text = "28天预测", width = 15, height = 2, command = arima_28)

    label_1.pack()
    label_2.pack()
    label_3.pack()
    label_4.pack()
    label_5.pack()
    button_4.pack()
    button_5.pack()
    button_6.pack()
    button_7.pack()
    button_8.pack()
    button_9.pack()
    button_10.pack()
    canvas.get_tk_widget().pack()
    
    top_des.mainloop()
```

```python
top = Tk()
top.geometry('1000x800')
top.title("股票数据分析平台")

label1 = Label(top, text="输入文件名:")
button1 = Button(top, text = "自动导入文件", width = 15, height = 2, command = auto_input_file)
text1 = Text(top,height=10, width=40)
button2 = Button(top, text = "手动导入文件", width = 15, height = 2, command = manu_input_file)
button3 = Button(top, text = "描述性统计", width = 15, height = 2, command = describe)
button_vis = Button(top, text = "查看数据框", width = 15, height = 2, command = vis_table)

button1.pack()
label1.pack()
text1.pack()
button2.pack()
button3.pack()
button_vis.pack()

top.mainloop()
```
