```python
from tkinter import*
import tkinter.messagebox as messagebox
import tkinter.font as tkFont
from tkintertable import TableCanvas, TableModel
import pandas as pd 
import numpy as np 
from sklearn import linear_model
import matplotlib.pyplot as plt
from matplotlib.figure import Figure 
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
from PIL import ImageGrab, Image, ImageTk
import pmdarima as pmd
from statsmodels.tsa.holtwinters import ExponentialSmoothing
%matplotlib inline 
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
# 自动导入文件函数
def auto_input_file():
    global SH
    SH = pd.read_csv("F:/大二.5下/课程/大计基/2020大计基/大作业/shanghai_index.csv")
    if(SH.shape[1] == 1):
        messagebox.showinfo("导入成功","单变量数据不能进行相关性分析！")
    else:
        messagebox.showinfo("导入成功","已导入全部数据")

# 手动导入文件函数
def manu_input_file():
    file_name = text1.get(1.0,END)
    global SH
    if(file_name[-4:-1] == 'csv'):
        SH = pd.read_csv("F:/大二.5下/课程/大计基/2020大计基/大作业/"+ str(file_name[0:-1]))
        if(SH.shape[1] == 1):
            messagebox.showinfo("导入成功","单变量数据不能进行相关性分析！")
        else:
            messagebox.showinfo("导入成功","已导入csv文件")
    elif(file_name[-4:-1] == 'txt'):
        SH = pd.read_table("F:/大二.5下/课程/大计基/2020大计基/大作业/"+ str(file_name[0:-1]))
        if(SH.shape[1] == 1):
            messagebox.showinfo("导入成功","单变量数据不能进行相关性分析！")
        else:
            messagebox.showinfo("导入成功","已导入txt文件")
    elif(file_name[-4:-1] == 'xls' or file_name[-5:-1] == 'xlsx'):
        SH = pd.read_excel("F:/大二.5下/课程/大计基/2020大计基/大作业/"+ str(file_name[0:-1]))
        if(SH.shape[1] == 1):
            messagebox.showinfo("导入成功","单变量数据不能进行相关性分析！")
        else:
            messagebox.showinfo("导入成功","已导入excel文件")
    else:
        messagebox.showinfo("导入失败","不支持的文件类型")

# 数据表可视化
def vis_table():
    frm = Frame()
    table = TableCanvas(frm)
    table.importCSV("F:/大二.5下/课程/大计基/2020大计基/大作业/shanghai_index.csv")
    table.show()
    frm.pack(fill=BOTH, expand=1)

# 绘出所有数据折线图
def draw_total():
    f_plot.clear()
    f_plot.plot(pd.to_datetime(SH.iloc[:,0]), SH.iloc[:,1])
    f_plot.set_xlabel("时间")
    f_plot.set_ylabel("股价")
    canvas.draw()

# 绘出线性回归结果
def draw_LR():
    f_plot.clear()
    regr = linear_model.LinearRegression()
    x = np.array(SH.iloc[:,1]).reshape(-1,1)
    y = np.array(SH.iloc[:,2]).reshape(-1,1)
    regr.fit(x, y)
    f_plot.scatter(x, y, color = 'y')
    xreg = np.linspace(2600, 3600, 10).reshape(-1,1)
    f_plot.plot(xreg, regr.predict(xreg), color = 'red', lw = 3)
    f_plot.set_xlabel("成交量")
    f_plot.set_ylabel("股价")
    canvas.draw()

# 绘出股票收益率条形图
def draw_revenue():
    f_plot.clear()
    SH["Revenue"] = (SH["Price"]-SH["Price"].shift(1))/SH["Price"].shift(1)
    f_plot.hist(SH["Revenue"])
    f_plot.set_xlabel("收益率")
    f_plot.set_ylabel("频数")
    canvas.draw()

# 保存当前图片
def save_fig():
    f.savefig("F:/大二.5下/课程/大计基/2020大计基/大作业/pic.png")
    messagebox.showinfo("提示","图片保存成功")

# ARIMA模型预测7天销量
def arima_7():
    f_plot.clear()
    model = pmd.auto_arima(SH["Price"][-120:], seasonal = True, m = 7, n_jobs = 2, suppress_warnings = True)
    forecasts = model.predict(7)
    f_plot.plot(pd.to_datetime(SH.iloc[:,0][-120:]), SH.iloc[:,1][-120:])
    f_plot.plot(pd.date_range(start='2020-03-12', periods = 7), forecasts, lw = 3)
    f_plot.set_xlabel("时间")
    f_plot.set_ylabel("股价")
    canvas.draw()

def arima_14():
    f_plot.clear()
    model = pmd.auto_arima(SH["Price"][-120:], seasonal = True, m = 7, n_jobs = 2, suppress_warnings = True)
    forecasts = model.predict(14)
    f_plot.plot(pd.to_datetime(SH.iloc[:,0][-120:]), SH.iloc[:,1][-120:])
    f_plot.plot(pd.date_range(start='2020-03-12', periods = 14), forecasts, lw = 3)
    f_plot.set_xlabel("时间")
    f_plot.set_ylabel("股价")
    canvas.draw()

def arima_28():
    f_plot.clear()
    model = pmd.auto_arima(SH["Price"][-120:], seasonal = True, m = 7, n_jobs = 2, suppress_warnings = True)
    forecasts = model.predict(28)
    f_plot.plot(pd.to_datetime(SH.iloc[:,0][-120:]), SH.iloc[:,1][-120:])
    f_plot.plot(pd.date_range(start='2020-03-12', periods = 28), forecasts, lw = 3)
    f_plot.set_xlabel("时间")
    f_plot.set_ylabel("股价")
    canvas.draw()

def Holt_winters_7():
    f_plot.clear()
    fits = ExponentialSmoothing(SH["Price"], trend = 'add', seasonal_periods=7).fit()
    forecasts = fits.forecast(7)
    f_plot.plot(pd.to_datetime(SH.iloc[:,0][-120:]), SH.iloc[:,1][-120:])
    f_plot.plot(pd.date_range(start='2020-03-12', periods = 7), forecasts, lw = 3)
    f_plot.set_xlabel("时间")
    f_plot.set_ylabel("股价")
    canvas.draw()

def Holt_winters_14():
    f_plot.clear()
    fits = ExponentialSmoothing(SH["Price"], trend = 'add', seasonal_periods=7).fit()
    forecasts = fits.forecast(14)
    f_plot.plot(pd.to_datetime(SH.iloc[:,0][-120:]), SH.iloc[:,1][-120:])
    f_plot.plot(pd.date_range(start='2020-03-12', periods = 14), forecasts, lw = 3)
    f_plot.set_xlabel("时间")
    f_plot.set_ylabel("股价")
    canvas.draw()

def Holt_winters_28():
    f_plot.clear()
    fits = ExponentialSmoothing(SH["Price"], trend = 'add', seasonal_periods=7).fit()
    forecasts = fits.forecast(28)
    f_plot.plot(pd.to_datetime(SH.iloc[:,0][-120:]), SH.iloc[:,1][-120:])
    f_plot.plot(pd.date_range(start='2020-03-12', periods = 28), forecasts, lw = 3)
    f_plot.set_xlabel("时间")
    f_plot.set_ylabel("股价")
    canvas.draw()
```

```python
def describe():
    top_des = Toplevel()
    top_des.geometry('800x800')
    top_des.title("描述性统计")

    global f
    f = Figure(figsize = (8,6), dpi = 100)
    global f_plot
    f_plot = f.add_subplot(111)
    global canvas
    canvas = FigureCanvasTkAgg(f, top_des)

    label_1 = Label(top_des, text="平均数: " + str(SH.iloc[:,1].mean()), font = ft3, fg = 'green')
    label_2 = Label(top_des, text="众数: " + str(np.argmax(np.bincount(SH.iloc[:,1]))), font = ft3, fg = 'green')
    label_3 = Label(top_des, text="中位数: " + str(SH.iloc[:,1].median()), font = ft3, fg = 'green')
    label_4 = Label(top_des, text="下四分位数: " + str(np.percentile(SH.iloc[:,1], 25)), font = ft3, fg = 'green')
    label_5 = Label(top_des, text="上四分位数: " + str(np.percentile(SH.iloc[:,1], 75)), font = ft3, fg = 'green')
    button_4 = Button(top_des, text = "A股总体趋势", width = 20, height = 2, command = draw_total, font = ft2, bg = 'white', fg = 'blue')
    button_5 = Button(top_des, text = "股价与成交量回归分析", width = 20, height = 2, command = draw_LR, font = ft2, bg = 'white', fg = 'blue')
    button_6 = Button(top_des, text = "A股收益率统计", width = 20, height = 2, command = draw_revenue, font = ft2, bg = 'white', fg = 'blue')
    button_7 = Button(top_des, text = "保存图片", width = 15, height = 2, command = save_fig, font = ft2, bg = 'green', fg = 'white')
    button_8 = Button(top_des, text = "7天预测", width = 20, height = 2, command = arima_7, font = ft2, bg = 'white', fg = 'red')
    button_9 = Button(top_des, text = "14天预测", width = 20, height = 2, command = arima_14, font = ft2, bg = 'white', fg = 'red')
    button_10 = Button(top_des, text = "28天预测", width = 20, height = 2, command = arima_28, font = ft2, bg = 'white', fg = 'red')
    button_HW1 = Button(top_des, text = "7天预测", width = 20, height = 2, command = Holt_winters_7, font = ft2, bg = 'white', fg = 'brown')
    button_HW2 = Button(top_des, text = "14天预测", width = 20, height = 2, command = Holt_winters_14, font = ft2, bg = 'white', fg = 'brown')
    button_HW3 = Button(top_des, text = "28天预测", width = 20, height = 2, command = Holt_winters_28, font = ft2, bg = 'white', fg = 'brown')
    label_arima = Label(top_des, text = "Auto.Arima股价预测:", font = ft4, fg = 'red')
    label_HW = Label(top_des, text = "Holt_Winters股价预测:", font = ft4, fg = 'brown')

    label_1.place(relx = 0.05, rely = 0.02)
    label_2.place(relx = 0.05, rely = 0.12)
    label_3.place(relx = 0.05, rely = 0.22)
    label_4.place(relx = 0.05, rely = 0.32)
    label_5.place(relx = 0.05, rely = 0.42)
    label_arima.place(relx = 0.47, rely = 0.28)
    label_HW.place(relx = 0.46, rely = 0.48)
    button_4.pack(anchor = 'e', padx = 30)
    button_5.pack(anchor = 'e', padx = 30)
    button_6.pack(anchor = 'e', padx = 30)
    button_8.pack(anchor = 'e', padx = 30)
    button_9.pack(anchor = 'e', padx = 30)
    button_10.pack(anchor = 'e', padx = 30)
    button_HW1.pack(anchor = 'e', padx = 30)
    button_HW2.pack(anchor = 'e', padx = 30)
    button_HW3.pack(anchor = 'e', padx = 30)
    button_7.pack(anchor = 'w', padx = 0)
    canvas.get_tk_widget().pack()
    
    top_des.mainloop()
```
